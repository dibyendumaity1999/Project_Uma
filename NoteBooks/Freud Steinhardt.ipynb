```python

def getFreud_ws(u):
    box = freud.box.Box.from_box(u.dimensions)
    data = u.select_atoms("name OW").positions
    system=(box, data)
    ql4 = freud.order.Steinhardt(4, wl = True, wl_normalize=True)
    ql4_ice = ql4.compute(system, neighbors={"num_neighbors": 15}).particle_order
    ql6 = freud.order.Steinhardt(6, wl = True, wl_normalize=True)
    ql6_ice = ql6.compute(system, neighbors={"num_neighbors": 15}).particle_order
    ql8 = freud.order.Steinhardt(8, wl = True, wl_normalize=True)
    ql8_ice = ql8.compute(system, neighbors={"num_neighbors": 15}).particle_order
    ql12 = freud.order.Steinhardt(12, wl = True, wl_normalize=True)
    ql12_ice = ql12.compute(system, neighbors={"num_neighbors": 15}).particle_order
 
    return ql4_ice, ql6_ice, ql8_ice, ql12_ice
```

```python

def getFreud_qs(u):
    box = freud.box.Box.from_box(u.dimensions)
    data = u.select_atoms("name OW").positions
    system=(box, data)
    ql4 = freud.order.Steinhardt(4)
    ql4_ice = ql4.compute(system, neighbors={"num_neighbors": 15}).particle_order
    ql6 = freud.order.Steinhardt(6)
    ql6_ice = ql6.compute(system, neighbors={"num_neighbors": 15}).particle_order
    ql8 = freud.order.Steinhardt(8)
    ql8_ice = ql8.compute(system, neighbors={"num_neighbors": 15}).particle_order
    ql12 = freud.order.Steinhardt(12)
    ql12_ice = ql12.compute(system, neighbors={"num_neighbors": 15}).particle_order

    return ql4_ice, ql6_ice, ql8_ice, ql12_ice
```
